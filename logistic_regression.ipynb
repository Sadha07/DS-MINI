```python
import numpy as np

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y = np.array([0, 0, 0, 1], dtype=float)

Xb = np.hstack([np.ones((X.shape[0], 1)), X])

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cost_function(theta, X, y):
    m = len(y)
    p = sigmoid(X @ theta)
    return (-1 / m) * np.sum(y * np.log(p) + (1 - y) * np.log(1 - p))

def gradient(theta, X, y):
    m = len(y)
    return (1 / m) * X.T @ (sigmoid(X @ theta) - y)

def gradient_descent(X, y, theta, alpha=0.1, num_iters=10000):
    for _ in range(num_iters):
        theta -= alpha * gradient(theta, X, y)
    return theta


theta0 = np.zeros(Xb.shape[1])
theta = gradient_descent(Xb, y, theta0, alpha=0.1, num_iters=10000)


probs = sigmoid(Xb @ theta)
preds = (probs >= 0.5).astype(int)


print("Learned theta:", theta)


accuracy = np.mean(preds == y) * 100
print(f"\nAccuracy: {accuracy:.2f}%")
```
